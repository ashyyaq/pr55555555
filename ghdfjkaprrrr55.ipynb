загружаем датасет и выводим его

```python
import pandas as pd

hsr = pd.read_csv(r"C:\Users\user\Downloads\hsr_character-data.csv")

hsr
```

датасет загрузился и вывелся

выводим размерность массива

```python
hsr.shape
```

вывелось количество строк и столбцов

выводим наименования колонок

```python
hsr.columns
```

вывелись наименивания колонок

выводим уникальные значения колонок

```python
for column in hsr.columns:
    unique_values = hsr[column].unique()
    print(f'{column}: {unique_values[:5]}')
```

вывелись уникальные значения колонок

сортируем по редкости персонажа и урону базовой атакой против мировых боссов и выводим отсортированный датасет

```python
sorted_hsr = hsr.sort_values(by=["rarity", "wb_b-atk"])

sorted_hsr
```

можно увидеть, что датасет отсортировался

удаляем ненужные столбцы 'path'(путь) и 'taunt'(насмешка) и выводим датасет уже с удаленными столбцами

```python
hsr = hsr.drop(["path", "taunt"], axis=1)

hsr
```

стоблцы удалились и датасет вывелся без них

заменяем пустые значения на значение 'unknown' в колонке 'character' и выводим изменный датасет

```python
hsr["character"].fillna("unknown", inplace=True)

hsr
```

пустые значения в колонке керектер заменились на анноун

удаляем дубликаты

```python
hsr.drop_duplicates(inplace=True)

hsr
```

дубликаты удалились

анализ с помощью info

```python
hsr.info()
```

info вывел количество непустых значений, типы данных каждого столбца и потребление памяти

анализ с помощью describe

```python
hsr.describe()
```

describe вывел среднее значение, стандартное отклонение, минимальное и максимальное значения, медиану и квартили

выборка данных с помощью loc по редкости персонажа 

```python
hsren = hsr.loc[hsr['rarity'] == 'SSR', ['character', 'rarity']]

hsren
```

вывелись колонки (их нет) после выборки

сохраняем датасет

```python
hsren.to_csv("hsr.csv", index=False)
```

обновленный датасет сохранился
